```python
from gym import envs
import numpy as np 


env = gym.make('Taxi-v2')
env._max_episode_steps = 40000
env.reset()

def policy_evaluation(policy, env, discount_factor, theta):
    
    V = np.zeros(env.nS)
    counter = 0
    while True: 
        counter += 1 
        delta = 0
        for s in range(env.nS): 
            vNew = 0
            for a in range(env.nA): 
                for prob, nextState, reward, done in env.P[s][a]:
                    vNew += policy[s][a] * prob * (reward + discount_factor * V[nextState])
                
            delta = max( delta, np.abs(V[s] - vNew))
            V[s] = vNew
        if delta < theta:
            break
                    

    return np.array(V)

def policy_iteration(env, discount_factor): 
    
    
    policy = np.ones([env.nS, env.nA]) / env.nA
    numItr = 0
    while True:
        numItr += 1 
        V = policy_evaluation(policy, env, discount_factor, 0.00001)
        policyStable = True 

        for s in range(env.nS):
            oldAction = np.argmax(policy[s])
           
            qValues = np.zeros( env.nA)
            for a in range(env.nA):
                for prob, nextState, reward, done in env.P[s][a]:
                    qValues[a] += prob * (reward + discount_factor * V[nextState])

            newAction = np.argmax(qValues)


            if oldAction != newAction:
                policyStable = False 
            policy[s] = np.zeros([env.nA])
            policy[s][newAction] = 1

        if policyStable:
            print(numItr)
            return policy, V
    return policy, np.zeros(env.env.S)

env.reset()
policyPI, valuePI = policy_iteration(env, 0.95)

print(policyPI)

print(valuePI)
```
